```python
country_name_mapping = {
    'Bolivia (Plurinational State of)': 'Bolivia',
    'Bosnia and Herzegovina': 'Bosnia-Herzegovina',
    'Brunei Darussalam': 'Brunei',
    "Côte d'Ivoire": "Cote d'Ivoire",
    "Democratic People's Republic of Korea": 'North Korea',
    'Democratic Republic of the Congo': 'DR Congo',
    'Guinea Bissau': 'Guinea-Bissau',
    'Iran (Islamic Republic of)': 'Iran',
    "Lao People's Democratic Republic": 'Laos',
    'Micronesia (Federated States of)': 'Micronesia',
    'Republic of Korea': 'South Korea',
    'Republic of Moldova': 'Moldova',
    'Saint Kitts and Nevis': 'St. Kitts and Nevis',
    'Saint Lucia': 'St. Lucia',
    'Saint Vincent and the Grenadines': 'St. Vincent and the Grenadines',
    'South Sudan': 'South Sudan',  #тут кстати хз. возможно этой страны как раз и нет в одном из датасетов
    'Syrian Arab Republic': 'Syria',
    'The former Yugoslav republic of Macedonia': 'Macedonia',
    'Timor Leste': 'Timor-Leste',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'United Republic of Tanzania': 'Tanzania',
    'United States of America': 'USA',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Viet Nam': 'Vietnam'
}

def standardize_country_name(name, mapping=None):
    name = name.replace('&', 'and')
    name = name.replace('_', ' ')
    name = ' '.join(name.split())

    if mapping:
        name = mapping.get(name, name)

    return name

life['CountryDisplay'] = life['CountryDisplay'].apply(lambda x: standardize_country_name(x, country_name_mapping))
drinks['country'] = drinks['country'].apply(lambda x: standardize_country_name(x, country_name_mapping))
```

Выведите минимальную, среднюю, медианную и максимальную оцененную продолжительность жизни людей от рождения (life expectancy at birth). Выберите любые три страны на ваш вкус и проделайте то же самое только для статей по этим странам. Что можно сказать на основе результатов, которые вы получили?

Я решил взять DisplayValue, так как в нем отображаются целочисленные значения (но можно взять и Numeric - разницы нет особо, я проверил)

```python
life_expect_birth = life[life['GhoDisplay'] == 'Life expectancy at birth (years)']

stats_functions = {
    'min': life_expect_birth['DisplayValue'].min,
    'mean': life_expect_birth['DisplayValue'].mean,
    'median': life_expect_birth['DisplayValue'].median,
    'max': life_expect_birth['DisplayValue'].max
}

for stat in stats_functions.keys():
    stat_value = stats_functions[stat]()
    print(f"{stat.capitalize()} life expectancy at birth: {stat_value}")
```

Ну можно тоже самое сделать, просто добавится массив из стран. Я решил взять самые дружелюбные и открытые страны: США, Япония, Франция)

```python
countries = ['USA', 'Japan', 'France']

stats_functions = {
    'min': 'min',
    'mean': 'mean',
    'median': 'median',
    'max': 'max'
}


for country in countries:
    life_expect_birth_country = life[(life['GhoDisplay'] == 'Life expectancy at birth (years)') &
                                     (life['CountryDisplay'] == country)]

    print(f"Статистики продолжительности жизни от рождения для {country}:")
    for stat_name, stat_func in stats_functions.items():
        stat_value = getattr(life_expect_birth_country['DisplayValue'], stat_func)()
        print(f"{stat_name.capitalize()}: {stat_value}")
    print("\n")
```

В целом, в Японии долго живут (оказывается это не стереотип даже) - по общей статистике даже азиаты живут дольше всех, во Франции чуть поменьше, а в США отсатют - с чем это связано, думаю столкнемся в дальнейшем

Я мог также забить вручную словарь, но если стран и функций очень много, то предположу, что есть изначально нам даны эти списки (в моем случае агр функции и три страны), и уже по ним автоматизировано прохожу циклом - если что, так проходить ими не запрещено, как и в предыдущих номерах

```python
countries = ['USA', 'Japan', 'France']
metrics = ['min', 'mean', 'median', 'max']

stats_data = {'Country': ['All'] + countries}

for metric in metrics:
    stats_data[metric] = []

life_expect_birth = life[life['GhoDisplay'] == 'Life expectancy at birth (years)']
for metric in metrics:
    func = getattr(life_expect_birth['DisplayValue'], metric)
    stats_data[metric].append(func())

for country in countries:
    country_data = life_expect_birth[life_expect_birth['CountryDisplay'] == country]
    for metric in metrics:
        func = getattr(country_data['DisplayValue'], metric)
        stats_data[metric].append(func())

result_df = pd.DataFrame(stats_data)

result_df.to_csv('life_expectancy_stats.csv', index=False)

result_df
```
